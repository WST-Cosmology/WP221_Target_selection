```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N_zm_bg_bright = np.load(f'LS_BG_BRIGHT_WST.npz')
N_zm_bg_faint = np.load(f'LS_BG_FAINT_WST.npz')
N_zm_elg = np.load(f'LS_ELG_WST.npz')
N_zm_lrg = np.load(f'LS_LRG_WST.npz')

N_zm_qso_no_H = np.load(f'COSMOS_QSO_WST_QSO_no_H.npz')
N_zm_qso_with_H = np.load(f'COSMOS_QSO_WST_QSO_with_H.npz')

N_zm_u = np.load(f'COSMOS_LBG_udropout.npz')
N_zm_u_highz = np.load(f'COSMOS_LBG_udropout_highz.npz')
N_zm_g = np.load(f'COSMOS_LBG_gdropout.npz')
N_zm_r = np.load(f'COSMOS_LBG_rdropout.npz')

samples_lbg = [N_zm_bg_bright, N_zm_bg_faint, N_zm_lrg,  N_zm_elg, N_zm_qso_no_H,N_zm_qso_with_H, N_zm_u_highz, N_zm_g, N_zm_r]

```

```python
mag_cut = [20.737125054200234, 21.6680344228211, 22.837125054200236, 25.337125054200236, 24.5, 24.5, 25.2, 25.2, 25.2]
name = ['bg_bright', 'bg_faint', 'lrg', 'elg', 'qso_no_H', 'qso_with_H', 'uLBGhighz', 'gLBG', 'rLBG']
for i, s in enumerate(samples_lbg):
    print(s['surface_deg2'])
    plt.plot(s['z_center'], np.cumsum(np.sum(s['object_count'][:,s['mag_center'] < mag_cut[i]], axis=1)/s['surface_deg2']), label = name[i])
plt.legend(ncols=3)
plt.xlabel('redshift')
plt.ylim(1)
plt.yscale('log')
plt.ylabel('density deg-2')
plt.ylim(100, 1e5)
```
