```python
data_path = '/global/cfs/cdirs/desi/users/cpayerne/data_WP221_Target_selection/photometric_catalogs/'
cosmos_clauds_hsc = Table.read(data_path + 'COSMOS_11bands-SExtractor-Lephare.fits')
```

```python
regions =  Regions.read(data_path + 'bright_stars_cosmos.reg')
```

```python
def selection_masks(ra, dec):
    skycoord = SkyCoord(ra, dec, unit='deg')
    in_mask=np.zeros(ra.size,dtype=bool)
    for i, reg in enumerate(regions):
        print(i)
        if i%1000==0: print(f'{i/len(regions):.1f}')
        %time in_mask += reg.contains(skycoord, mywcs)
        
    return in_mask
```

```python
in_mask = selection_masks(cosmos_clauds_hsc['RA'], cosmos_clauds_hsc['DEC'])
```
